```python
total_adj = df.groupby(["TIPO DE BENEFICIARIO","SUBDIRECCION","PROGRAMA","SEXO"])["ESTADO_RESOLUCION_CONCURSO"].value_counts().unstack(3).fillna(0).astype(int)
total_adj
```

```python
help(map)
```

```python
#Operaciones matemáticas con columnas
total_adj["TOTAL"] = total_adj["HOMBRE"] + total_adj["MUJER"] + total_adj["SIN INFORMACION"] 
total_adj["HOMBRE (%)"] = total_adj["HOMBRE"]/total_adj["TOTAL"]*100
total_adj["MUJER (%)"] = total_adj["MUJER"]/total_adj["TOTAL"]*100
total_adj["S/I (%)"] = total_adj["SIN INFORMACION"]/total_adj["TOTAL"]*100
total_adj["HOMBRE (%)"] = total_adj["HOMBRE (%)"].map('{:,.2f}%'.format)
total_adj["MUJER (%)"] = total_adj["MUJER (%)"].map('{:,.2f}%'.format)
total_adj["S/I (%)"] = total_adj["S/I (%)"].map('{:,.2f}%'.format)
total_adj # http://www.dipres.gob.cl/597/articles-290515_doc_pdf.pdf
```

## 5.2 Analisis de la Subdirección de Capital Humano

Ahora aplicaremos un filtro a los datos, para quedarnos solo con los concursos gestionados por la SCH durante el período 2016 - 2020.

### 5.2.1 Datos de la SCH

image.png

```python
df_sch = df[(df["SUBDIRECCION"] == "CAPITAL HUMANO")]
df_sch
```

### 5.2.2 Datos de la SCH por concurso y estado de adjudicación.

Ahora agruparemos los datos para todos los concuros de cada año académico.

```python
resumen = pd.DataFrame(df_sch.groupby(["AÑO DE FALLO","NOMBRE DEL CONCURSO","ESTADO_RESOLUCION_CONCURSO"])["CODIGO DEL PROYECTO"].count())
resumen
```

### 5.2.3 Datos de la SCH para concursos/beneficios adjudicados. 

image.png

image.png

Desde aqui en adelante, trabajaremos solo con los datos de los que tentan un "ESTADO_RESOLUCION_CONCURSO" ADJUDICADO.

```python
df_sch_ocde = df[(df["SUBDIRECCION"] == "CAPITAL HUMANO") & (df["ESTADO_RESOLUCION_CONCURSO"] == "ADJUDICADO")] # Filtro para adjudicados
df_sch_ocde 
```

### 5.2.4 Resultados por ÁREA OCDE (adjudicados)

Verifiquemos el comportamiento de los concursos/beneficios otorgados segùn Área OECD.

```python
df_sch_ocde.groupby(["DISCIPLINA OECD"])["CODIGO DEL PROYECTO"].count()
```

Desglosemos el resultado anterior para verificar la adjudicación total por año.

```python
df_sch_adj = df_sch_ocde.groupby(["AÑO DE FALLO","DISCIPLINA OECD"])["CODIGO DEL PROYECTO"].count().unstack(1)
df_sch_adj 
```

### 5.2.5 Resultados por Sexo (adjudicados)

Revisemos ahora como se comporta la adjudicación al verificar los resultados por sexo.

```python
df_sch_ocde.groupby(["AÑO DE FALLO","SEXO"])["CODIGO DEL PROYECTO"].count()
```

Veamos ahora como se comporta por concurso y año de fallo.

```python
adjudicados_sexo = pd.DataFrame(df_sch_ocde.groupby(["AÑO DE FALLO","NOMBRE DEL CONCURSO"])["SEXO"].value_counts().unstack(2))
adjudicados_sexo
```

Calculemos los porcentajes y veamos las brechas globales por concurso.

```python
adjudicados_sexo["TOTAL"] = adjudicados_sexo["HOMBRE"] + adjudicados_sexo["MUJER"]
adjudicados_sexo["HOMBRE (%)"] = adjudicados_sexo["HOMBRE"]/adjudicados_sexo["TOTAL"]*100
adjudicados_sexo["MUJER (%)"] = adjudicados_sexo["MUJER"]/adjudicados_sexo["TOTAL"]*100
adjudicados_sexo["HOMBRE (%)"] = adjudicados_sexo["HOMBRE (%)"].map('{:,.2f}%'.format)
adjudicados_sexo["MUJER (%)"] = adjudicados_sexo["MUJER (%)"].map('{:,.2f}%'.format)
adjudicados_sexo
```

Elaboremos un gráfico para visualizar los resultados.

```python
grafico = df_sch_ocde.groupby("AÑO DE FALLO")["SEXO"].value_counts().unstack(1)
#df_sch_ocde.groupby(["SEXO"])["AÑO DE FALLO"].value_counts().unstack(0).plot.bar()
```

### 5.2.6 Resultados por Macro Zona (adjudicados)

Revisemos la adjudicaciòn por Macro Zona.

```python
df_sch_mz = df_sch_ocde.groupby(["AÑO DE FALLO"])["MACRO ZONA"].value_counts().unstack(1).fillna(0).astype(int)
df_sch_mz 
```

# 6 Exportar Dataframes a Excel.

```python
#Exportar a Excel

df0 = df.copy() # todos los datos 2016-2020
df1 = total_adj.copy() # Total por subdirecciones
df2 = df_sch_ocde.copy() # datos SCH ADJUDICADOS
df3 = resumen.copy() # SCH resumen por año y concurso (estados de adjudicación)
df4 = df_sch_adj.copy() # SCH resumen de adjudicados por AREA OCDE
df5 = adjudicados_sexo.copy() # SCH resumen de adjudicados por SEXO
df6 = df_sch_mz.copy() # SCH resumen de adjudicados por MACRO ZONA

with pd.ExcelWriter('ANID Git-Hub (2016-2020).xlsx') as writer:  
    df0.to_excel(writer, sheet_name='Datos 2016-2020')
    df1.to_excel(writer, sheet_name='Total Adjudicación 2016 - 2020')
    df2.to_excel(writer, sheet_name='SCH ADJ (2016-2020)')
    df3.to_excel(writer, sheet_name='SCH por año y concurso (ADJ)')
    df4.to_excel(writer, sheet_name='SCH por AREA OCDE (ADJ)')
    df5.to_excel(writer, sheet_name='SCH por SEXO (ADJ)')
    df6.to_excel(writer, sheet_name='SCH por MACRO ZONA (ADJ)')
```
